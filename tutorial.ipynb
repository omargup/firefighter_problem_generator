# Creating instances of the Moving Firefighter Problem

```python
import networkx as nx
import numpy as np

import movingfp.gen as mfp

np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})
```

## Erdos instance

```python
n = 8
p = 0.5
dim =  2
fighter_pos = None
num_fires =  2
connected = True
seed = 3210

x = mfp.erdos(n, p, dim, fighter_pos, num_fires, connected, seed)
mfp.draw_mfp(x)
```

```python
# Adjacency matrix (Fire graph)
x.A
```

```python
# Distance matrix (Firefighter graph)
# Last row and column are firefighter distances to every node
x.D
```
